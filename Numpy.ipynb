# Numpy

Python lists are : 
 - Flexible but slowfor numerical operations
 - Not designed for matrix math or
 - Does not support vectorised operation
 - Are memory inefficient for large numerical data

Why use numpy?
- Fast than pure python for numerical computation
- Vectorized operation (no explicit loops)
- Broadcasting (operate on arrays of different shape)
- Memory efficient
- Comes with powerful linear algebra, random tools etc

Use of numpy :
- Scientific python

```python
import numpy as np
```

### 3. np.linspace(start, stop, num=x)
stop - inclusive

```python
np_linespace_1=np.linspace(1,100,4)
print(np_linespace_1)
print(type(np_linespace_1))
```

```python
np_linespace_2=np.linspace(1,10,10)
print(np_linespace_2)
print(type(np_linespace_2))
```

```python
for i in np.linspace(1,10,10):
    print(i)
```

```python
for i in np.linspace(1,10,10,dtype=np.int8):
    print(i)
```

### 4. np.zeros(shape)

```python
npz_1=np.zeros(5)
npz_1
```

```python
npz_2=np.zeros((5,5),dtype=np.int8)
print(npz_2)
```

```python
npz_2=np.zeros((5,5))
print(npz_2)
```

```python
npz_3=np.zeros((5,5,5))
print(npz_3)
```
